## Beta Calculations

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-darkgrid')
%matplotlib inline
import pandas as pd

import warnings
warnings.simplefilter("ignore")

from data import get_stock_data

start_date = '2016-1-1'
end_date = '2021-9-17'

data = pd.DataFrame()
data['NESTLEIND.NS'] = get_stock_data('NESTLEIND.NS',start_date,end_date)['Close']
data['^NSEI'] = get_stock_data('^NSEI',start_date,end_date)['Close']

data.tail(2)
```

$$ 
    Stock\ returns_i = \alpha + \beta * Market\ returns_i
$$

```python
data_pc = data.pct_change().dropna()
plt.figure(figsize=(10,7))
plt.scatter(data_pc['NESTLEIND.NS'],data_pc['^NSEI'], color='b')
plt.xlabel('Nestle India')
plt.ylabel('Nifty')
plt.show()
```

```python
import statsmodels.api as sm
line = sm.OLS(data_pc['NESTLEIND.NS'],data_pc['^NSEI'])
line = line.fit()
beta = line.params[0]
beta
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data_pc['NESTLEIND.NS'],data_pc['^NSEI'], color='b')
x = data_pc['NESTLEIND.NS']
plt.plot(x, beta*x, color='r')
plt.xlabel('NESTLEIND')
plt.ylabel('Nifty')
plt.show()
```
