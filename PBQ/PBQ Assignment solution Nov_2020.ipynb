**Q1: Write a customized function to calculate the factorial of a number. Do not use any inbuilt functions. Create your own.**

```python
# To ask for an input from the user
num = int(input("Enter a number - "))

# The required function
fact = 1

def factorial(num):
    fact = 1
    if num < 0:
        print("Facorial can not be found for negative number")
    else :
        for i in range(1, num+1):
            fact = fact * i
        print("The factorial of {0} is {1}".format(num, fact))
        return fact

# Calling the function for the inputted value    
factorial (num)
```

**Q2: Write a Python program which iterates through a sequence of 50 randomly generated integers between 100 and 200, and derive two different series/arrays/lists from it, one which contains even multiples of 3 and the other which contains odd multiples of 5.**

**Hint: You have to use the random submodule of the NumPy library and the '%' operator in Python.**

```python
import numpy as np

# Using np.random.randint() to generate the random integers
data = np.random.randint(100, 201, 50)

even_multiples_3=data[(data%3==0) & (data%2==0)]
odd_multiples_5=data[(data%5==0) & (data%2!=0)]

print(data)
print("-------------------------------------------------------------------------")
print("even multiples of 3 -- ",even_multiples_3)
print("odd multiples of 5 -- ",odd_multiples_5)
```

**Q3: Write a Python program to generate a sequence of 1000 numbers drawn from a standard uniform distribution and plot it on a chart. Compute its mean. Use the np.where clause to create a sequence of only numbers greater than 0.75 from the list. Plot a chart of those numbers. Check how many such numbers exist. Is it on expected lines? Comment. Please use np.random.seed(1234) at the beginning of your code.**

 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Using np.random.seed(1234) and np.random.uniform() to create 1000 random samples from a standard uniform distribution
np.random.seed(1234)
data1 = np.random.uniform(0,1,1000)

# plotting data
plt.figure(figsize=(30,7))
plt.plot(data1)
plt.show()
print("Mean of distribution -- ",np.mean(data1))
```

```python
# Using np.where to filter all the index locations at which the data value exceeds 0.75 
data2 = np.where(data1 > 0.75)
# Fetching all values greater than 0.75
data3 = data1[data2]
# plotting all values greater than 0.75
plt.figure(figsize=(30,7))
print("Number exists -- ", len(data3))
plt.plot(data3)
plt.show()
```

Comment: Yes, this is on expected lines. As we have generated a random sample of size 1000 from a uniform distribution, we expect around a quarter of the total value(near 250)to be be greater than 0.75
