```python
import pandas as pd
import matplotlib.pyplot as plt  
%matplotlib inline
plt.style.use('seaborn-darkgrid')

l = [9,13,8,14,10,12,7,15,8,12]
df = pd.DataFrame(l)
df.columns = ['price']
df.plot(figsize=(10,7), color='b')
plt.show()
```

```python
df['change_in_price'] = df.price - df.price.shift(1)
df['price_tm1'] = df.price.shift(1)
```

```python
plt.figure(figsize=(10,7))
# scatter(X,Y)
df = df.dropna()
x=df['price_tm1']
y=df['change_in_price']
plt.scatter(x,y)
plt.ylabel('Change in Price')
plt.xlabel('Price (t - 1)')
plt.plot()
plt.show()
```
